### Construct a frequency histogram for bell.jpg, and plot the histogram using plt.bar (search on google for syntax)

### Perform thresholding on bell.jpg using the above constructed histogram. 

```python
img_bgr = cv2.imread('./bell.jpeg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_bgr = cv2.imread('./bell.jpeg')
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

threshold = np.median(img_gray)
new_img = np.zeros_like(img_gray)
new_img += (img_gray >= threshold)
new_img *= 255
plt.imshow(new_img, cmap="gray")
```

### Plot the RGB histograms for the image lotus.jpg using inbuilt functions. (Look up cv2.calcHist)



```python
img_bgr = cv2.imread('./lotus.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img_bgr],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

### Perform thresholding on fingerprint.jpg and find the threshold. 



```python
img_bgr = cv2.imread('./fingerprint.jpg')
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

threshold = np.mean(img_gray)
new_img = np.zeros_like(img_gray)
new_img += (img_gray >= threshold)
new_img *= 255
plt.imshow(new_img, cmap="gray")
```

### Challenge: Try chroma keying of images as a homework. Take fg.jpg foreground image and put it on bg.jpg. 

```python
img_bgr = cv2.imread('./bg.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_bgr = cv2.imread('./einstein.jpg')
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap="gray")
print(img_gray.shape)
```

```python
rows, cols = img_gray.shape
new_img = np.copy(img_gray)
sq_size = 3
for r in range(rows - (sq_size - 1)):
    for c in range(cols - (sq_size - 1)):
        med = np.median(img_gray[r:r+sq_size,c:c+sq_size])
        new_img[r:r+sq_size, c:c+sq_size] = med
plt.imshow(new_img, cmap="gray")
```

```python
img_bgr = cv2.imread('./coins.png')
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
img_gray = np.array(img_gray, dtype=int)
plt.imshow(img_gray, cmap="gray")
print(img_gray.shape)
```

```python
rows, cols = img_gray.shape
new_img = np.copy(img_gray)
sq_size = 3
laplacian = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])
for r in range(rows - (sq_size - 1)):
    for c in range(cols - (sq_size - 1)):
        add = laplacian * img_gray[r:r+sq_size, c:c+sq_size]
        new_img[(r+r+sq_size)//2, (c+c+sq_size)//2] += np.mean(add)
plt.imshow(new_img, cmap="gray")
```

```python
img_bgr = cv2.imread('./bike.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

new_img = np.copy(img_rgb)

for i in range(img_rgb.shape[0]):
    for j in range(img_rgb.shape[1]):
        r, g, b = img_rgb[i][j]
        outputRed = min((r * .393) + (g *.769) + (b * .189), 255)
        outputGreen = min((r * .349) + (g *.686) + (b * .168), 255)
        outputBlue = min((r * .272) + (g *.534) + (b * .131), 255)
        
        new_img[i][j] = np.array([outputRed, outputGreen, outputBlue])

plt.imshow(new_img)
effect_img = new_img
```

```python
tex_bgr = cv2.imread('./texture2.jpg')
tex_rgb = cv2.cvtColor(tex_bgr, cv2.COLOR_BGR2RGB)

dim = (img_rgb.shape[1], img_rgb.shape[0])
tex_re = cv2.resize(tex_rgb, dim)
plt.imshow(tex_re)
```

```python
wt = 0.5
new_img = wt * effect_img + (1 - wt) * tex_re
new_img = effect_img * (tex_re / 255)
new_img = np.array(new_img, dtype = int)

plt.imshow(new_img)
```
